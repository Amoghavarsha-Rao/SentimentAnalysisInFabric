```python
df = spark.read.option("multiline", "true").json("Files/BingResults.json")
# df now is a Spark DataFrame containing JSON data from "Files/BingResults.json".
display(df) 
```

```python
df=df.select("value")
#dropping all values except for the json_object column that has the attributes of the news article
display(df)
```

```python
from pyspark.sql.functions import explode
df1 = df.select(explode(df["value"]).alias("json_object"))
#using explode to separate the values into an RDD each
display(df1)
```

```python
newJSON = df1.toJSON().collect()
print(newJSON[1])
```

```python
import json
finalJSON=json.loads(newJSON[1])
print(finalJSON)
```

```python
print(finalJSON['json_object']['description'])
#drilled down and got the description in the first entry in finalJSON
```

```python
print(finalJSON["json_object"]["name"])
print(finalJSON["json_object"]["description"])
print(finalJSON["json_object"]["category"])
print(finalJSON["json_object"]["url"])
print(finalJSON["json_object"]["image"]["thumbnail"]["contentUrl"])
print(finalJSON["json_object"]["provider"][0]["name"])
print(finalJSON["json_object"]["datePublished"])
#this gets the values of the attributes in the first article
```

```python
title = []
description = []
category = []
image = []
url = []
provider = []
datePublished = []
#creating lists that will store extracted values in the order that they are exctracted from the JSON

for json_str in newJSON:
    try:
        article = json.loads(json_str)

        
        title.append(article["json_object"]["name"])
        description.append(article["json_object"]["description"])
        category.append(article["json_object"]["category"])
        url.append(article["json_object"]["url"])
        image.append(article["json_object"]["image"]["thumbnail"]["contentUrl"])
        provider.append(article["json_object"]["provider"][0]["name"])
        datePublished.append(article["json_object"]["datePublished"])

    except Exception as e:
        print(f"error processing object, value doesn't exist: {e}")
        
```

```python
print(datePublished)
```

```python
from pyspark.sql.types import StructType, StructField, StringType
cleanedData = list(zip(title, description,category, url,image, provider,datePublished))
schema = StructType([
    StructField("title",StringType(),True),
    StructField("description",StringType(),True),
    StructField("category",StringType(),True),
    StructField("url",StringType(),True),
    StructField("image",StringType(),True),
    StructField("provider",StringType(),True),
    StructField("datePublished",StringType(),True),
])

cleanedDf = spark.createDataFrame(cleanedData, schema=schema)
```

```python
display(cleanedDf)
```

```python
from pyspark.sql.functions import to_date, date_format

cleanedDf = cleanedDf.withColumn("datePublished", date_format(to_date("datePublished"),"dd-MM-yyyy"))
```

```python
display(cleanedDf.limit(5))
```

```python
from pyspark.sql.utils import AnalysisException

try:
    
    table_name = "revisedtable"
    
    print("Table created or replaced")

except AnalysisException:
    
    print("Table already exists")

       
    cleanedDf.createOrReplaceTempView("vw_df_category_final")

    spark.sql(f"""
            MERGE INTO {table_name} target_table
            USING vw_df_category_final source_view
                ON source_view.url = target_table.url OR
                source_view.category = target_table.category OR
                source_view.description = target_table.description OR
                source_view.image_url = target_table.image_url OR
                source_view.provider = target_table.provider OR
                source_view.date_published = target_table.date_published 
                WHEN MATCHED THEN 
                    UPDATE SET *
                WHEN NOT MATCHED THEN 
                    INSERT *
            """)
```
